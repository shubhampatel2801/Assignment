## Apriori Algorithm

#### 1. Assiciation rules with 20% Support and 60% confidence

```python
# with 20% support

Frequent_itemsets=apriori(Book,min_support=0.2,use_colnames=True,)
Frequent_itemsets
```

```python
# with 60% confidence

Rules=association_rules(Frequent_itemsets,metric='lift',min_threshold=0.6)
Rules
```

```python
# Lift ratio > 1 is a good infliential rule in selecting the associaated transactions

Rules[Rules.lift>1]
```

```python
# Visulization of obtained rule

plt.scatter(Rules['support'],Rules['confidence'])
plt.xlabel('support')
plt.ylabel('confidence')
plt.show
```

### 2. Association rules with 5% Support and 80% confidence

```python
# with 5 % support 

Frequent_itemsets2=apriori(Book,min_support=0.05,use_colnames=True)
Frequent_itemsets2
```

```python
# 90% confidence

Rules2=association_rules(Frequent_itemsets2,metric='lift',min_threshold=0.8)
Rules2
```

```python
# Lift ratio > 1 is a good infliential rule in selecting the associaated transactions

Rules2[Rules2.lift>1]
```

```python
# Visulization of obtained rule

plt.scatter(Rules2['support'],Rules2['confidence'])
plt.xlabel('support')
plt.ylabel('confidence')
plt.show
```
